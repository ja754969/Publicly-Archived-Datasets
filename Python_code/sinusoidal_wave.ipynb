```python
%matplotlib inline
import os 
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2.5,2.5,100) # measured in meters
x.shape
Amp = 2 # Amplitude, the peak deviation of the function from zero.
f = 0.5   # f, ordinary frequency, the number of oscillations (cycles) that occur each second of time.
omega = 2*np.pi*f   #  = 2f, angular frequency, the rate of change of the function argument in units of radians per second

 = 1 #  phase angle (), specifies (in radians) where in its cycle the oscillation is at t = 0.
# When  is non-zero, the entire waveform appears to be shifted in time by the amount / seconds. 
# A negative value represents a delay, and a positive value represents an advance.
y = Amp*np.sin(omega*x-) # Angle, in radians (2 rad equals 360 degrees)

plt.plot(x,y)
plt.grid(color='gray', linestyle=':', linewidth=2)
plt.axis([-2.5, 2.5, -2.5, 2.5])
plt.xticks(np.arange(-2.5, 2.5, step=0.5))
```
